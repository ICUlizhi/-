#### (c)
定义:
- $f(A)=U_p^*(A)-U_q^*(A)$
- $U_x^*(A)=\underset{y_x}\max \, U_x(y_x)$
- $y_x\in [0,1]$为选择资产$x$时,资产$x$的占比

注意到,
$$
\begin{align*}
U_x&=E(r)-A\sigma^2\\
&=r_f+y_x(E(r_x)-r_f)-\frac{1}{2}Ay_x^2\sigma_x^2\\
\frac{dU_x}{dy_x}&=E(r_x)-r_f-A\sigma_x^2y_x\\
\end{align*}
$$
因此,
$$\frac{dU_x}{dy_x}>0\Leftrightarrow 0<y_x<\min\{\frac{E(r_x)-r_f}{A\sigma_x^2},1\}$$
也即
$$y_x^*=\underset{y_x}{\argmax}\,U_x=\min\{\frac{E(r_x)-r_f}{A\sigma_x^2},1\}$$
代入得,
$$
y_p=\min\{\frac{5}{A},1\},y_q=\min\{\frac{1.6}{A},1\}
$$
$$
\begin{align*}
f(A)&=
\begin{cases}
&\frac{21A}{800}-\frac{1}{20} &0<A\leq1.6\\
&\frac{1}{20}-\frac{A}{200}-\frac{2}{25A} &1.6<A\leq 5\\
&\frac{9}{200} &5<A
\end{cases}
\end{align*}
$$
解 $f(A)<0$ 得 $A<2$ , 也即当 $A<2$ 时,投资者会选择 $Q$ 而不是 $P$ 来和短期国库券构建投资组合

Variance of returns is given by,
$$
\begin{align*}
var(R_i)&=var(\alpha_i+\beta_iR_m+e_i)\\
&=\beta_i^2\sigma_m^2+\sigma^2(e_i)
\end{align*}
$$
Percentage of idiosyncratic risk is given by,
$$
\frac{var(e_i)}{var(R_i)}
$$

#### (b)
The portfolio return is 
$$R_p=\frac{1}{5}\sum_{i=1}^5\alpha_i+\beta_iR_m+e_i$$
Variance of this return is given by
$$
\begin{align*}
var(R_p)&=var(\frac{1}{5}\sum_{i=1}^5\alpha_i+\beta_iR_m+e_i)\\
&=var(\frac{1}{5}\sum_{i=1}^5\beta_iR_m)+var(\frac{1}{5}\sum_{i=1}^5e_i)\\
&=(\frac{1}{5}\sum_{i=1}^5\beta_i)^2\sigma_m^2+\sum_{i=1}^5(\frac{1}{5})^2\sigma^2(e_i)\\
&=0.00698
\end{align*}
$$
Percentage of idiosyncratic risk is given by 
$$\frac{var(\frac{1}{5}\sum_{i=1}^5e_i)}{var(R_p)}=\frac{0.004}{0.00698}=57.33\%$$

## 5
#### (a)
To construct an arbitrage portfolio without borrowing, consider a portfolio $S$ which consists of weight 0.5 in $Q$ and 0.5 in bills. The beta of portfolio $S$ is
$$\beta_S = 0.5\beta_Q = 1$$
which is the same as $\beta_P$ . The expected return of S is
$$E(r_S ) = 0.5E(r_Q) + 0.5r_f = 13\%$$
which is greater than the expected return of $P$ . Hence
$$r_S = E(r_S) + \beta_SF = 13\% + F$$
Selling $P$ and using the proceeds to buy S gives a return $rS − rP = 2\%$ regardless of $F$.
#### (b)
The condition for no arbitrage is
$$r_P=r_S\Rightarrow E(r_Q)=(E(r_P)-r_f)\beta_Q/\beta_P+r_f=20\%$$
#### (c)
The sharpe ratio is given by:
$$S=\frac{E(P)-r_f}{\sigma_P}=8\%$$
#### (d)
$$
\begin{align*}
&\frac{E(P)-r_f}{\beta_P}=\frac{E(r_S)-r_f}{\beta_S}\\
\Rightarrow&E(r_S)=8\%\beta_S+4\%
\end{align*}
$$

### Types of Financial Assets
- **Fixed-income Securities (debt securities)(固定收益证券):** Promises a fixed stream of income or a stream of income determined by a specified formula
    - issued by anyone who borrows money
    - bond investor is known as creditor, bond issuer is known as debtor
    - government bond, corporate bond, financial bond, etc
    - less risky and its asset price is least closely tied to the financial condition of the issuers
- **Equity (common stock)(股本):** : Represents ownership share in a corporation
  - each share of common stock gives its holders an equal share in the ownership of the firm, the same amount of profits, and one vote on the matter of corporate governance
  - investors hold equity in anticipation(预期) of dividend and capital gain(股息和资本收益) from appreciation
  - equity investor is known as stockholder
  - common stock represents a residual claim(剩余索取权) on the assets of a corporation.
  - if proceeds(收益) from the sale of all its assets are not enough to pay off the debts, the creditors(债权人) cannot ask the stockholders for more money to meet the shortfall(不足). This is known as **limited liability**.

Financial assets allow us to make the most of the economy’s real assets.
- the information role
- consumption timing
- allocation of risk
- separation of ownership and management

## L2 Allocating Resource Over Time

###  Discounted Cash Flow
**Interest rate(i)** is a promised rate of return
> Why are interest rates usually positive?
- People care more about today than future
- The least one can do is to hold money
- In the aftermath of the Financial Crisis, Euro Zone, Switzerland, Denmark, Sweden and Japan allowed rates to fall to slightly below 0%
- Negative rate policy diminishes the liquidity of funds.

### Compounding
- present value (PV)
$$FV_n=PV_0(1+i)^n$$
- discount factor : $\frac{1}{(1+i)^n}$
- A useful tool in analyzing the timing of cash flows is known as a time line

### Annuity and Perpetuity
An **annuity** is a series of equal cash flows received or paid at regular intervals over a specified period of time. The cash flows can be received or paid at the beginning or end of each period.

The present value of an annuity can be calculated using the following formula:

$$
PV = \frac{C}{(1 + r)} + \frac{C}{(1 + r)^2} + \ldots + \frac{C}{(1 + r)^n}
$$
$$FV_t = \frac{C}{i}[(1+i)^t-1]$$

An **ordinary annuity** is an annuity where the cash flows occur at the end of each period. The formula for calculating the present value of an ordinary annuity is the same as the formula mentioned above.

A **perpetuity** is an annuity that continues indefinitely, with equal cash flows received or paid at regular intervals. The present value of a perpetuity can be calculated using the following formula:

$$
PV = \frac{C}{r}
$$

### APR and EFF
**Annual percentage rate (年度百分率)**
$$APR = \text{period rate}\times \text{number of periods per year(m)}$$
**Effective annual rate (有效年利率)**
$$EAR = (1 + \text{period rate})^{m} - 1$$
Holding $APR$ fixed, $EAR$ goes to $e^{APR} −1$ as $m$ goes to infinity

### T-Bills
- US debt instruments with maturities of one year or less
- Low-risk investment with a broad and liquid secondary market
- T-bills are free of default risk (违约风险) and generally have lower interest rate than corporate issues of comparable securities
- Interest earned on Treasuries is exempt from state and local taxes

#### T-Bill Auctions(拍卖)
- The **face value** of a T-Bill is the amount received by holder on the maturity(到期) date
- Auction participant submit **bids(出价)** immediately before the auction
  - Investors submit menus(菜单) of competitive bids by size(规模) and yield(收益率,其实是债券价格).
  - Primary dealers(一级交易商) are required to participate in every auction.[List]
  - Small (≤ 5 million) non-competitive bids are always accepted.
- Treasury accepts the lowest yields up until the announced
quantity is allotted in full and rejects the remaining bids.
  - Market clearing (市场出清) rate: stop-out bids at the highest winning yield are prorated.
- Auction results are released within minutes of the closing time.
  - All successful bidders receive the bond at the same stop-out price.
  
#### T-Bill Yields
- The difference between the purchase price and the face value of the T-bill is the investor’s return
- Two-commonly mathematical formulas:
1. The Discount Yield Method
- Takes into account the return as a percent of the face value (rather
than its purchase price)
- The discount method tends to understate the yield
- Competitive bids are discount rates
- The high rate (in the set-up) is the market clearing discount rate
$$\text{Discount Yield}=\frac{FV-P}{FV}\times\frac{360}{M}$$
1. The Investment Yield Method
-  Relates to the investor’s return to the purchase price of the bill
$$\text{Investment Yield}=\frac{FV-P}{P}\times\frac{365or366}{M}$$


### Fixed Income Capital Market
- Longer term debt securities (1-30 years)
> Examples : Treasury Notes & Bonds,Corporate Bonds,International Bonds,Municipal Bonds,...
- Interest paid periodically (coupon payments (息票))
- Semiannual(半年) payments
- Sold at auction (like bills)

#### Bond pricing
$$P_B=\sum_{t=1}^T\frac{C_t}{(1+r)^t}+\frac{\text{Face Value}}{(1+r)^T}$$
- Coupon rate (息票率) is fixed for the life of the bond
  - The stream of payments are fixed (e.g. $125 every 6 months)
  - While the bond price can vary over time
- For any given (market) bond price, the bond yield is the interest rate implicit in the payment stream
- The **yield** (yield to maturity, 到期收益率,r) of a bond is defined as the interest rate at which the present value of the payment stream equals the bond price

> What drives Yields?<br/>
> Time value,Inflation,Creait risk,Liquidity,Taxes

### Equities(产权)
Common Stock
- Ownership
- Shares in a company confer(授予) voting rights at shareholders’ meetings, and a **residual claim** (剩余索取权) on assets
- Shareholders receive earnings paid as **dividends** (股利) an potential capital gains (资本收益).

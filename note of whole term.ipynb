### 3 Annuity and Perpetuity
An **annuity** is a series of equal cash flows received or paid at regular intervals over a specified period of time. The cash flows can be received or paid at the beginning or end of each period.

The present value of an annuity can be calculated using the following formula:

$$
PV = \frac{C}{(1 + r)} + \frac{C}{(1 + r)^2} + \ldots + \frac{C}{(1 + r)^n}
$$
$$FV_t = \frac{C}{i}[(1+i)^t-1]$$

An **ordinary annuity** is an annuity where the cash flows occur at the end of each period. The formula for calculating the present value of an ordinary annuity is the same as the formula mentioned above.

A **perpetuity** is an annuity that continues indefinitely, with equal cash flows received or paid at regular intervals. The present value of a perpetuity can be calculated using the following formula:

$$
PV = \frac{C}{r}
$$

### 4 APR and EFF
**Annual percentage rate (年度百分率)**
$$APR = \text{period rate}\times \text{number of periods per year(m)}$$
**Effective annual rate (有效年利率)**
$$EAR = (1 + \text{period rate})^{m} - 1$$
Holding $APR$ fixed, $EAR$ goes to $e^{APR} −1$ as $m$ goes to infinity

### 3 T-Bills
- US debt instruments with maturities of one year or less

#### T-Bill Auctions(拍卖)
- The **face value** of a T-Bill is the amount received by holder on the maturity(到期) date
- Auction participant submit **bids(出价)** immediately before the auction
  - Investors submit menus(菜单) of competitive bids by size(规模) and yield(收益率,其实是债券价格).
  - Primary dealers(一级交易商) are required to participate in every auction.[List]
  - Small (≤ 5 million) non-competitive bids are always accepted.
- Treasury accepts the lowest yields up until the announced
quantity is allotted in full and rejects the remaining bids.
  - Market clearing (市场出清) rate: stop-out bids at the highest winning yield are prorated.
- Auction results are released within minutes of the closing time.
  - All successful bidders receive the bond at the same stop-out price.
  
#### T-Bill Yields
- The difference between the purchase price and the face value of the T-bill is the investor’s return
- Two-commonly mathematical formulas:
1. The Discount Yield Method
- Takes into account the return as a percent of the face value (rather
than its purchase price)
- The discount method tends to understate the yield
- Competitive bids are discount rates
- The high rate (in the set-up) is the market clearing discount rate
$$\text{Discount Yield}=\frac{FV-P}{FV}\times\frac{360}{M}$$
1. The Investment Yield Method
-  Relates to the investor’s return to the purchase price of the bill
$$\text{Investment Yield}=\frac{FV-P}{P}\times\frac{365or366}{M}$$

### 2 The Trading Process

**Types of markets**
- 直接搜索：买家和卖家直接寻找对方
- 经纪市场：经纪人(Brokers) 搜索买家和卖方
    - 一级市场就是一个例子。
- 经销商市场：经销商(Dealers) 拥有购买和出售资产的库存
- 拍卖市场：交易者聚集在一个地方进行交易

**Bid & Ask Prices**
- 两个公布的价格：最佳出价和最佳要价
- 买卖价差(Bid-ask spread)

**Market Orders**
- 按最高价买入或按最高价卖出
- “在市场”买入股票或“在市场上”卖出股票
- 仅指定数量
- 立即执行
- 交易者收到当前市场价格
- 发布的要价可能因此而变化
> Potential issues?
- 大额市场订单可能以多种价格完成
  - 公布的报价代表了交易一定数量股票的承诺。
- 其他交易员首先行动
  - 另一个交易员可以击败我们的投资者获得报价。我们的订单执行价格会更低
- 价格突然波动

**Limit Orders**
- 一种价格或有指令（另一种是停止指令）
- 投资者可以同时指定价格和数量下订单。
- 愿意购买的最高价格（限购）或愿意出售的最低价格（限售）
- 限价订单是买卖价格的来源
- 记录在限额订单簿中的所有限额订单
- 每个订单（市场或限价）都会改变账面


### 3 Buying on Margin

- 债务融资资产购买：投资者投入一些资金,剩余部分向经纪人借款
- 经纪人即期贷款(Broker’s call loans): 投资者借入部分
- 从经纪人那里购买股票的价格
  - 随借随贷利率
- 股票作为抵押品(collateral)
- 实例
  - 投资者有6000美元
  - 以每股100美元的价格购买100股
  - 从经纪人那里借了4000美元
- 投资者必须满足保证金要求(margin requirements)

- **Margin:** 投资者出资购买价款的比例；剩余部分是从经纪人那里借来的。两种类型的保证金要求：
1. **Initial margin**（适用于购买时,通常为50%）
2. **Maintenance margin**（随后适用,对价格变动敏感）

若价格下跌幅度过大,则违反保证金要求,投资者将获得追加保证金通知(margin call)

## L6 Funds(基金)

### 1 Types of Investment Fund
- **开放式基金**（例如共同基金、对冲基金）是最大的类别,由基金创建和赎回的股票。
  - 投资者从基金购买股票,然后按照每个交易日收盘时确定的资产净值将其卖回给基金。
- **封闭式基金**在证券交易所上市交易,固定数量的已发行股票（需额外发行）。
  - 封闭式基金份额以市场价格在投资者之间进行交易,通常不等于基金的资产净值。
- **交易所交易基金**（ETF）是最新的基金类型,结合了开放式基金和封闭式基金的最佳特点。
  - 发明人之间在交易所进行股票交易,但存在套利机制（套利机制）比保持市场价格的机制接近资产净值


#### Net Asset Value
$$\text{Net Asset Value}=\frac{\text{Market value of assets }-\text{Liabilities}}{\text{Shares Outstanding}}$$

### 2 开放式基金

- 共同基金是开放式基金中最大的一类,允许投资者每天购买或赎回股票。
  - 投资者与该基金进行交易,而不是相互交易,也没有做空。
  - 对冲基金是开放式的,但提供月度/季度流动性。
- 股份可按资产净值购买或赎回（不得二级市场交易）在交易日结束时计算
- 购买和赎回不影响资产净值
  - 投资者以21美元的资产净值购买10000股基金
  - 净资产增加210000美元（现金）
  - 流通股增加10000股
  - 资产净值（以前）=42000000/2000000=21美元
  - 资产净值（后）=42210000/2010000=21美元（不变）
- 基金种类：货币市场、股票、行业、债券、指数等。

**投资者向指数基金和ETF的迁移**
- 大多数共同基金都是积极管理的(actively managed), 用一个投资组合经理选择证券,但选择指数基金( index funds)正在变得越来越受欢迎。
- 由于人们认为积极的投资组合管理不会增加价值,投资者已从主动型基金转向被动型基金
- 指数基金提供更低的费用,确保投资者盈利基准收益。

### 3 封闭式基金
- 封闭式基金（CEF）投资于资产组合,但与基金不同的是,CEF发行的股票在交易所交易。
  - CEF通过IPO筹集资金,然后由投资组合经理进行投资。
  - CEF可以在IPO后发行更多股票,但这种情况并不常见。
  - 股票是不可赎回的——通过出售给其他投资者来退出。
- 要投资封闭式基金,你需要从另一个投资者那里购买股票。
  - 价格由市场供求决定,不必等于基金每天公布的净资产价值(NAV)。
- 股价可能（而且经常）与资产净值不同。
  - 如果价格>NAV,则表示该基金的交易溢价(premium)
  - 如果价格<NAV,则基金交易折价(discount)
  - 溢价=（价格-资产净值）/资产净值
- 封闭式基金的资产净值通常有持续的折扣。

**封闭式基金套利**
> 投资者有可能从CEF折价中套利吗？
- 积极的方法是迫使管理层清算(清算)基金,套利者可以通过收购大部分股份来做到这一点。
  - 清算将迫使该基金的价格与其资产价值持平。
  - 这导致了较大的交易成本,如果投资者预期清算,他们会要求净资产值（NAV）,这会消除套利利润。
- 被动的做法是买入CEF股票,对冲资产风险。
  - 做空基金的所有持股可能很困难(流动性不足)。
  - 可行的替代方案是以折扣价购买多元化的CEF投资组合,并通过做空广泛的股票指数进行对冲。

### 4 交易所交易基金（ETF,Exchange Traded Funds）
- ETF(交易所交易基金) 结合两者的特点
  - 可以在基金到期前赎回
  - 可以在二级市场上自由交易
- 大多数ETF都是指数基金,也就是说,它们复制某些股票市场指数（例如标准普尔500指数）或债券市场指数。

**潜在优势：**
- 像股票一样连续交易,投资者在当天交易股票,而不是在一天结束时赎回
- 可以卖空或以保证金购买
- 成本较低,因为它们没有得到积极管理

**潜在缺点：**
- 价格可能偏离NAV（在较小程度上）
- 必须从经纪人处购买

### 1 HPR : holding period return ($i$)
$$r=\frac{1+i}{1+\pi}-1\approx i-\pi$$
- 名义回报率nominal return $i$
- 实际回报率real return $r$
- 通胀率 inflation rate $\pi$
$$r_{t+1}=\frac{P_{t+1}-P_t}{P_t}-\pi_{t+1}$$
- Returns are displayed on logarithmic scale
- 短期国债分布并不反映证券本身的风险，而更多地反映了随着时间推移，无风险利率的变化


### 2 Historical risk and returns
- 我们定义了**风险溢价(risk premium)** 作为风险资产的预期HPR与无风险利率（国债、货币市场基金）之间的差额。
- 标准差被用作市场波动的指标，因此也被用作风险的指标
- 在任何特定时期，风险资产的实际回报率与无风险回报率之间的差异称为**超额回报(excess return)** .

## L10 Constructing Portfolios
BKM: Chapter 6.2, 6.4-6.5
### 1 Combining assets
For a a portfolio(投资组合):
$$E(r_p)=w_1E(r_1)+w_2E(r_2)$$
$$\sigma_p^2=w_1^2\sigma_1^2+w_2^2\sigma_2^2+2w_1w_2\sigma_{12}$$
- If $\sigma_1=0$,$\sigma_{12}=0$

假设投资组合C由risk-free的部分和投资组合P组成,则
$$E(r_c)=(1-y)r_f+yE(r_p)=r_f+y(E(r_p)-r_f)$$
- $y$ : weight in risky asset
- $E(r_p)-r_f$ : risk-premium of P 
$$\sigma_c=y\sigma_p$$

#### Sharpe ratio
一个风险资产/组合的**夏普收益波动性(Reward-to-Volatility(Sharpe)Ratio)** 为:
$$S=\frac{\text{Risk premium}}{\text{SD of excess returns}}=\frac{E(r_p)-r_f}{\sigma_p}$$



### 2 Capital Allocation Line (CAL)
资本配置线(CAL):
$$E(r_c)=r_f+S\sigma_c$$
- Linear relation between $E(r_c)$ and $\sigma_c$ , intercept $r_f$ and slope $S$
- An investor can choose any point on the CAL by selecting the
corresponding portfolio weight $y$
  - $y=0$ : only bills are chosen , $E(r_c)=r_f,\sigma_c=0$
  - $0<y<1$ : both assets are chosen
  - $y=1$ : only risky assets are chosen , $E(r_c)=E(r_p),\sigma_c=\sigma_p$
  - $y>1$ : investor has more that 100% of wealth in risky asset (this
represents borrowing at $r_f$ to buy on margin)

### 3 Optimal choice
$$
\begin{align*}
\frac{dU}{dy}&=\frac{E(r)-\frac{1}{2}A\sigma^2}{y}\\
&=r_f-E(r_p)-yA\sigma_p^2\\
y^*&=\frac{E(r_p)-r_f}{A\sigma_p^2}
\end{align*}
$$


## L11 Optimal Risky Portfolios
### 1 Diversification and portfolio risk
- Market risk
  - Market-wide的风险来源
  - 即使在分散化(Diversification)之后仍然存在
  - 也称为：系统性(Systematic) 或Nondiversifiable
- Firm-specific risk
  - 可通过分散化消除的风险
  - 也称为：Nonsystematic 或 Diversifiable

### 2 Portfolio of two risky assets
bonds & equities<br/> 
$E(r_d),E(r_e),\sigma_d,\sigma_e,\sigma_{de}=\rho_{de}\sigma_d\sigma_e,w_e=x,w_d=1-x$
- **投资组合可行集(portfolio opportunity set)**: $\{(E(r_p),\sigma_p)|x\}$
- **最小方差组合(mimimum variance portfolio)** : $\frac{d\sigma_p^2}{dx}=0$
  - 方差低于债券
  - 预期回报高于债券
  - 股票的积极权重
  
在只有两种风险资产可选时，投资者会选择位于**上斜段(upward-sloping segment)** 的投资组合，因为这些组合提供了更高的预期收益率，虽然伴随着更高的风险。这被称为**有效边界、有效前沿(efficient frontier)**，因为它代表了在给定风险水平下能够实现最优收益的投资选择。最终的选择将取决于投资者的偏好，即他们对风险和收益的权衡。

## L13 Index Models
BKM: Chapter 8.1-8.3

### Concepts
- **系统(systematic)风险** 来自于货币政策、贸易政策、汇率、消费者信心等
  - 这种风险影响到所有资产，而且是不可分散的
- **特殊(idiosyncratic)风险** 来自专利、诉讼、合并、管理决策、人员更替等
  - 这种风险是不系统的、具体的和多样化的
- **excess return(超额收益率)** $R_i=r_i-r_f$<br/>
### Index model
Consider a market index M with $R_m=r_m-r_f$, index models are base on the hypothesis that movements in $R_i$ and $R_m$ are correlated.
$$R_i = \alpha_i+\beta_iR_m +e_i$$
- $e_i$ 是反映特定风险的随机变量
- $\beta_i$ 反映对影响所有证券的因素的敏感性(可能是大的、小的甚至是负的)
  - 航空公司、建筑公司的贝塔系数很高
  - 公用事业、黄金的贝塔系数较低/为负
$$E(e_i)=0,cov(e_i,R_m)=0,cov(e_i,e_j)=0$$

### Linear regression
回归后(T observations):
- 残差 $e_i=R_i-(\alpha_i+\beta_iR_m)$
- $m=\sum_{i=1}^T e_i = 0$
- $S^2(e_i)=\frac{1}{T-2}\sum_{t=1}^{T}e_i^2$

### Variance Decomposition
$$\sigma_i^2= \beta_i^2\sigma_m^2+\sigma^2(e_i) $$
$$\sigma^2_p=\frac{1}{n^2}\sum \sigma^2_i \rightarrow 0$$

## L14 Capital Asset Pricing Model
BKM: Chapter 9.1
### Assumptions
- 个体行为：
  - 投资者相对于市场规模较小
  - 投资者都理性(理性的)((mean-variance optimizers)
  - 所有投资者都有同质的(homogeneous)预期
  - 所有投资者都有一个共同的规划期限
市场结构：
  - 资产交易公开
  - 无税or手续费
  - 借款利率等于无风险利率

#### Capital Market Line
Suppose the market portfolio is $M$,we have $E(r_m),\sigma_m$
- CML : CAL that is constructed from the risk-free asset and the market portfolio


### Individual Securities
$$\{w_i\}_n,\{\sigma_{ij}\}_{n\times n}$$
- Covariance of $r_i$ with $r_m$ is :
$$\sigma_{im}=\sum_{j=1}^{n}w_i\sigma_{ij}$$
- Variance of Market Portfolio :
$$\sigma_m^2=\sum_{i=1}^n\sum_{j=1}^n w_iw_j\sigma_{ij}=\sum_{i=1}^n w_i\sigma_{im}$$
- Share of $\sigma_m^2$ attribuable to asset $i$ is 
$$\frac{w_i\sigma_{im}}{\sigma_m^2}$$
- Share of $E(r_m)-r_f$ attribuable to asset $i$ is 
$$\frac{w_i(E(r_i)-r_f)}{E(r_m)-r_f}$$

> **Key idea of CAPM** : Assets which add a lot to the risk of the market portfolio should also add a lot to the return
$$
\begin{align*}
&\frac{w_i(E(r_i)-r_f)}{E(r_m)-r_f}=\frac{w_i\sigma_{im}}{\sigma_m^2}\\
\Rightarrow & E(r_i) = r_f+\frac{\sigma_{im}}{\sigma_m^2}(E(r_m)-r_f)\quad(\text{SML: }\beta-E(r))
\end{align*}
$$
$$\beta_i = \frac{\sigma_{im}}{\sigma_m^2},\beta_m=1$$

- CAPM 提供了识别错误定价证券的方法
  - 预期收益与其预测值之间的差额是资产的$\alpha$,可正可负
  - 投资者可以通过增持正$\alpha$证券和减持负$\alpha$证券来获得更好的风险收益组合
  - 在均衡状态下，证券应该没有$\alpha$,这意味着它们位于SML上.CAPM是基于均衡(而非套利)
- CAPM 和 Index Model的关系显而易见 (same $\beta$ ,CAPM predicts $\alpha = 0$)
- **Efficient Market Hypothesis**(有效市场假说)

## L15 Arbitrage Pricing Theory 
BKM: Chapter 10.1-10.2
### Two key ideas
- 特有风险可以通过多样化有效地消除
- 资产价格应该不存在套利机会


### Returns
- Individual Securities
  - 联立 Index model , $R_i$定义, $F=r_m-E(r_m)$ 可得:
  $$r_i=E(r_i)+\beta_iF+e_i$$
- Portfolio
  $$r_i=E(r_i)+\beta_iF$$

> The APT states that for all well-diversiﬁed portfolios, the risk premium is proportional to beta:
$$\frac{E(r_P)-r_f}{\beta_P}=k,\forall P$$

### APT and CAPM
- 在APT中令$k=E(r_m)-r_f$即可得到CAPM
- 但APT适用于多元化程度较高的投资组合,不针对个股,个股不必躺在SML上.
- APT不需要限制性的CAPM假设,而且是基于套利而不是均衡.

## Corporate bonds
企业通过发行债券来借钱。一些在纽约证券交易所债券平台上交易，大多数在场外交易(柜台交易、场外交易)。
- **可赎回债券(Callable bonds)：** 可在到期日之前回购
  - 发行人(firm)可以在赎回日延长或注销债券。为什么？
  - 发行者可以选择“赎回(all)”(注销retire)高票面利率债务，并发行新的低票面利率债券，以减少利息支付--称为“退款(refunding)”
- **可转换债券(Convertible bonds)：** 可兑换公司普通股
- **可卖出债券(Puttable bonds)：** 给予债券持有人(投资者)选择退休或延长债券寿命的选择。
  - 如果债券的票面利率太低，债券持有人可以决定不延期，而是收回本金。

- **浮动利率债券(Floating-rate bonds)：** 票面利率可调。
  - 受制于市场利率的一般水平，而不受特殊的firm条件制约

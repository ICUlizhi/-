## 1
#### (a) 计算每种情况下的概率及对应的抵押品池的回报

1. **两只债券都违约**
   - 概率：0.2 * 0.2 = 0.04
   - 回收金额：0.4 * $15M + 0.6 * $25M = $6M + $15M = $21M

2. **两只债券都存续**
   - 概率：0.8 * 0.8 = 0.64
   - 回收金额：$15M + $25M = $40M

3. **A违约B存续**
   - 概率：0.2 * 0.8 = 0.16
   - 回收金额：0.4 * $15M + $25M = $6M + $25M = $31M

4. **A存续B违约**
   - 概率：0.8 * 0.2 = 0.16
   - 回收金额：$15M + 0.6 * $25M = $15M + $15M = $30M

预期回报计算：
$$ E[\text{回报}] = 0.04 \times 21 + 0.64 \times 40 + 0.16 \times 31 + 0.16 \times 30 $$
$$ E[\text{回报}] = 0.84 + 25.6 + 4.96 + 4.8 = 36.2 $$

所以，抵押品池的预期回报是 $36.2M。

#### (b) 高级债券的预期回报（以百分比表示）

- 高级债券吸收前$10M损失后的剩余回报。
- 各种情况的回报：
  - 两只债券都违约：$21M - $10M = $11M
  - 两只债券都存续：$40M - $10M = $30M
  - A违约B存续：$31M - $10M = $21M
  - A存续B违约：$30M - $10M = $20M

预期回报计算：
$$ E[\text{高级债券回报}] = 0.04 \times 11 + 0.64 \times 30 + 0.16 \times 21 + 0.16 \times 20 $$
$$ E[\text{高级债券回报}] = 0.44 + 19.2 + 3.36 + 3.2 = 26.2 $$

所以，高级债券的预期回报是 $26.2M，占其本金的百分比为：
$$ \frac{26.2}{30} \times 100\% \approx 87.33\% $$

#### (c) 次级债券的预期回报（以百分比表示）

- 次级债券吸收前$10M损失后的剩余回报。
- 各种情况的回报：
  - 两只债券都违约：$10M（由于损失超过$10M）
  - 两只债券都存续：$10M
  - A违约B存续：$10M
  - A存续B违约：$10M

预期回报计算：
$$ E[\text{次级债券回报}] = 10 $$

所以，次级债券的预期回报是 $10M，占其本金的百分比为：
$$ \frac{10}{10} \times 100\% = 100\% $$

#### (d) 违约概率完全相关时的预期回报（以百分比表示）

- 如果违约概率完全相关，那么只有两种情况：
  1. **两只债券都违约**：
     - 概率：0.2
     - 回收金额：$21M
     - 高级债券：$21M - $10M = $11M（占本金的 36.67%）
     - 次级债券：$10M（占本金的 100%）

  2. **两只债券都存续**：
     - 概率：0.8
     - 回收金额：$40M
     - 高级债券：$30M（占本金的 100%）
     - 次级债券：$10M（占本金的 100%）

预期回报计算：
$$ E[\text{高级债券回报}] = 0.2 \times 11 + 0.8 \times 30 = 2.2 + 24 = 26.2 $$
$$ \text{占本金的百分比} = \frac{26.2}{30} \times 100\% \approx 87.33\% $$

$$ E[\text{次级债券回报}] = 10 $$
$$ \text{占本金的百分比} = \frac{10}{10} \times 100\% = 100\% $$

所以，当违约概率完全相关时：
- 高级债券的预期回报是 87.33%
- 次级债券的预期回报是 100%

## 4

假设股票当前的价格为 $ S_0 = 100 $，6 个月后和12 个月后的价格如图所示。年化无风险收益率为 4%，这意味着半年的无风险收益率为 $ r = \frac{4\%}{2} = 2\% $。

#### (a) 计算看涨期权的价格

首先，我们需要计算6个月后的期权价格，然后推导到当前价格。

##### 1. 计算12个月后各节点的期权价格
看涨期权行权价 $ K = 97 $。

- $ S_{++} = 121 $，行权价为 97，收益为 $ \max(121 - 97, 0) = 24 $。
- $ S_{+-} = 100 $，行权价为 97，收益为 $ \max(100 - 97, 0) = 3 $。
- $ S_{--} = 81 $，行权价为 97，收益为 $ \max(81 - 97, 0) = 0 $。

##### 2. 计算6个月后各节点的期权价格

假设6个月后的价格是 $ S_+ = 110 $ 和 $ S_- = 90 $。

期权价格 $ C_+ $ 和 $ C_- $ 可以通过折现计算得到：

$$ C_+ = \frac{1}{1 + r} \left( q C_{++} + (1 - q) C_{+-} \right) $$
$$ C_- = \frac{1}{1 + r} \left( q C_{+-} + (1 - q) C_{--} \right) $$

其中，$ q $ 为风险中性概率：

$$ q = \frac{e^{0.04} - d}{u - d} $$

设 $ u = 1.1 $，$ d = 0.9 $，计算 $ q $：

$$ q = \frac{e^{0.04/2} - 0.9}{1.1 - 0.9} \approx \frac{1.02 - 0.9}{0.2} = 0.6 $$

所以：

$$ C_+ = \frac{1}{1.02} \left( 0.6 \times 24 + 0.4 \times 3 \right) = \frac{1}{1.02} \left( 14.4 + 1.2 \right) = \frac{15.6}{1.02} \approx 15.29 $$

$$ C_- = \frac{1}{1.02} \left( 0.6 \times 3 + 0.4 \times 0 \right) = \frac{1}{1.02} \left( 1.8 \right) = \frac{1.8}{1.02} \approx 1.76 $$

##### 3. 计算当前期权价格

当前期权价格 $ C_0 $ 可以通过折现计算得到：

$$ C_0 = \frac{1}{1.02} \left( 0.6 \times 15.29 + 0.4 \times 1.76 \right) $$
$$ C_0 = \frac{1}{1.02} \left( 9.174 + 0.704 \right) = \frac{9.878}{1.02} \approx 9.69 $$

所以，看涨期权的当前价格约为 $9.69。

#### (b) 计算看跌期权的价格

使用期权平价公式 $ C - P = S_0 - Ke^{-rT} $，求看跌期权价格。

已知：

- $ C = 9.69 $
- $ S_0 = 100 $
- $ K = 97 $
- $ r = 0.04 $
- $ T = 1 $ 年

所以：

$$ 9.69 - P = 100 - 97e^{-0.04} $$
$$ P = 9.69 - 100 + 97e^{-0.04} $$
$$ P \approx 9.69 - 100 + 97 \times 0.9608 $$
$$ P \approx 9.69 - 100 + 93.1976 $$
$$ P \approx 2.89 $$

所以，看跌期权的当前价格约为 $2.89。

#### (c) 考虑股利的看涨期权价格

假设12个月内会支付每股 $2 的股利。在行权时，股价调整为 $98（假设股利不影响股价波动）。

##### 1. 计算调整后的期权价格

各节点的期权价格调整后为：

- $ S_{++} = 119 $，行权价为 97，收益为 $ \max(119 - 97, 0) = 22 $。
- $ S_{+-} = 98 $，行权价为 97，收益为 $ \max(98 - 97, 0) = 1 $。
- $ S_{--} = 79 $，行权价为 97，收益为 $ \max(79 - 97, 0) = 0 $。

##### 2. 计算6个月后各节点的期权价格

同理，计算新的 $ C_+ $ 和 $ C_- $：

$$ C_+ = \frac{1}{1.02} \left( 0.6 \times 22 + 0.4 \times 1 \right) = \frac{1}{1.02} \left( 13.2 + 0.4 \right) = \frac{13.6}{1.02} \approx 13.33 $$

$$ C_- = \frac{1}{1.02} \left( 0.6 \times 1 + 0.4 \times 0 \right) = \frac{1}{1.02} \left( 0.6 \right) = \frac{0.6}{1.02} \approx 0.59 $$

##### 3. 计算当前期权价格

当前期权价格 $ C_0 $ 可以通过折现计算得到：

$$ C_0 = \frac{1}{1.02} \left( 0.6 \times 13.33 + 0.4 \times 0.59 \right) $$
$$ C_0 = \frac{1}{1.02} \left( 7.998 + 0.236 \right) = \frac{8.234}{1.02} \approx 8.07 $$

所以，考虑股利后的看涨期权价格约为 $8.07。

#### (d) 考虑股利的看跌期权价格

使用期权平价公式 $ C - P = S_0 - Ke^{-rT} $，求看跌期权价格。

已知：

- $ C = 8.07 $
- $ S_0 = 100 $
- $ K = 97 $
- $ r = 0.04 $
- $ T = 1 $ 年

所以：

$$ 8.07 - P = 100 - 97e^{-0.04} $$
$$ P = 8.07 - 100 + 97e^{-0.04} $$
$$ P \approx 8.07 - 100 + 93.1976 $$
$$ P \approx 1.27 $$

所以，考虑股利后的看跌期权价格约为 $1.27。

#### (a) 雪球资产的年化收益率与标的资产价格变化的关系

雪球资产的年化收益率随着标的资产价格的变化与以下哪个更接近？
- A. 买入看涨期权
- B. 卖出看涨期权
- C. 买入看跌期权
- D. 卖出看跌期权

由于雪球资产在标的资产价格上涨至敲出价格以上时合约立刻停止，并按照承诺收益率支付本息，这种情况类似于卖出看涨期权。当价格跌破敲入价格且未重新上涨至敲出价格时，投资者面临损失，这种情况也类似于卖出看涨期权。因此，答案是：
- **B. 卖出看涨期权**

#### (b) 利用套利定价理论计算雪球资产的承诺收益率 $i\%$

假设标的资产价格每年变化一次，每次变化有上涨 25% 和下跌 20% 两种可能。初始价格为 $100，敲出价格为 $103，敲入价格为 $75，存续期为 3 年。

根据给定的条件：

- 若标的资产在任一年结束时达到或超过 $103，雪球资产停止，并支付承诺收益率的本息。
- 若标的资产在任一年结束时跌至或低于 $75 且未再反弹至 $103，3 年期满时仅支付标的资产期末价格。
- 若标的资产在 3 年内未达到 $103 且未跌破 $75，3 年期满时支付承诺收益率的本息。

我们设定：

- $ S_0 = 100 $
- $ S_u = 100 \times 1.25 = 125 $
- $ S_d = 100 \times 0.8 = 80 $

根据无套利定价模型，假设风险中性概率 $ q $：

$$ q = \frac{e^{r_f} - d}{u - d} = \frac{1 - 0.8}{1.25 - 0.8} = \frac{0.2}{0.45} \approx 0.4444 $$

不考虑无风险利率 $ r_f = 0\% $，一年后可能的价格情况：

- $ S_u = 125 $
- $ S_d = 80 $

期望收益需要满足：

$$ V_0 = \frac{1}{1+r} [q \times V_u + (1 - q) \times V_d] $$

假设承诺收益率为 $ i $，计算一年后的收益：

- 如果 $ S_u \geq 103 $，支付 $ 100 \times (1 + i) = 100(1 + i) $
- 如果 $ S_d < 103 $，价格为 $ 80 $

计算3年的情况（考虑期望值）：

- $ V_0 = 100 $

通过计算，承诺收益率 $ i $ 可以通过方程求解得出。计算如下：

$$ 100 = \frac{1}{1+0} [0.4444 \times 100(1 + i) + 0.5556 \times 80] $$

简化：

$$ 100 = 0.4444 \times 100(1 + i) + 0.5556 \times 80 $$
$$ 100 = 44.44 + 44.44i + 44.448 $$
$$ 11.112 = 44.44i $$
$$ i = \frac{11.112}{44.44} \approx 0.25 $$

所以，承诺收益率 $ i \approx 25\% $。

#### (c) 投资机构的对冲行为对市场的影响

假设投资机构出售了一份雪球资产，他不会直接通过雪球和购买者对赌，而是会选择买入标的资产对冲掉自己的风险。详细分析如下：

(i) 当标的资产价格上涨时，投资机构会买入标的资产：

如果标的资产价格上涨到敲出价格 $103 以上，雪球资产会立刻停止，投资机构需要支付承诺收益率的本息。为了对冲这一风险，投资机构会在标的资产价格接近 $103 时买入标的资产，从而确保其可以支付给投资者。

这种买入行为会导致标的资产价格进一步上涨。

所以答案是：**A. 买入**，**C. 上涨**

(ii) 当标的资产价格下降，但没有跌破敲入价时，投资机构会卖出标的资产：

如果标的资产价格下降，但没有跌破敲入价格 $75，雪球资产不会停止，投资机构仍需对冲其风险。为了减少持有标的资产的成本，投资机构可能会选择卖出部分标的资产。

这种卖出行为会导致标的资产价格进一步下跌。

所以答案是：**B. 卖出**，**D. 下跌**

(iii) 当标的资产价格下降并跌破敲入价时，投资机构会买入标的资产：

如果标的资产价格下降并跌破敲入价格 $75，投资机构面临较大的风险，因为雪球资产可能在期满时仅支付标的资产期末价格。为了对冲这一风险，投资机构可能会在标的资产价格低于敲入价时买入标的资产，以期未来价格反弹至敲出价以上。

这种买入行为会导致标的资产价格上涨。

所以答案是：**A. 买入**，**C. 上涨**

| 条件   | 行为      | 结果     |
|--------|-----------|----------|
| (i)    | A. 买入   | C. 上涨  |
| (ii)   | B. 卖出   | D. 下跌  |
| (iii)  | A. 买入   | C. 上涨  |

#### (d) 通过以上分析可以看到：

情况 **(i)** 降低了标的资产的波动，情况 **(ii)** 加剧了标的资产的波动。今年年初的股灾导致各个 A 股市场指数大幅下跌，以这些指数为标的资产的雪球类似于情况 **(ii)**。

## 6

#### (a) 如果不存在套利机会，计算 F0

根据无套利定价原理，远期汇率 $ F_0 $ 的计算公式为：

$$ F_0 = S_0 \times \frac{(1 + r_{USD})}{(1 + r_{GBP})} $$

已知：
- 美元的一年期无风险利率 $ r_{USD} = 3\% $
- 英镑的一年期无风险利率 $ r_{GBP} = 8\% $
- 美元兑英镑的即期汇率 $ S_0 = 1.25 $ 美元/英镑

所以：

$$ F_0 = 1.25 \times \frac{(1 + 0.03)}{(1 + 0.08)} $$
$$ F_0 = 1.25 \times \frac{1.03}{1.08} $$
$$ F_0 \approx 1.25 \times 0.9537 $$
$$ F_0 \approx 1.1921 $$

所以，如果不存在套利机会，计算得到的远期汇率 $ F_0 $ 约为 1.1921 美元/英镑。

#### (b) 计算含手续费情况下 F0 的取值范围

如果远期合约在兑现时要缴纳 1% 的手续费，远期汇率的取值范围需要考虑手续费的影响。

手续费使得远期汇率的实际价格 $ F_0' $ 必须在以下范围内：

$$ F_0 \times (1 - 0.01) \leq F_0' \leq F_0 \times (1 + 0.01) $$

因此：

$$ 1.1921 \times 0.99 \leq F_0' \leq 1.1921 \times 1.01 $$
$$ 1.1802 \leq F_0' \leq 1.2040 $$

所以，含手续费情况下远期汇率 $ F_0' $ 的取值范围为 1.1802 至 1.2040 美元/英镑。

#### (c) 如果 F0 = 1.22，描述套利策略

假设 F0 = 1.22。我们可以利用套利策略来获利。假设 1 年后的美元兑英镑即期汇率记为 $ S_1 $。

##### 1. 当期操作：
- 借入 1 英镑（GBP），需在 1 年后偿还 $ 1 \times (1 + 0.08) = 1.08 $ 英镑。
- 将借来的 1 英镑按即期汇率 $ S_0 = 1.25 $ 兑换成美元，得到 1.25 美元。
- 将 1.25 美元以 3% 的利率存入美元账户，1 年后得到 $ 1.25 \times (1 + 0.03) = 1.2875 $ 美元。

##### 2. 签订远期合约：
- 签订 1 年后按远期汇率 $ F_0 = 1.22 $ 将 1.2875 美元兑换成英镑的远期合约。

##### 3. 一年后操作：
- 收到 1.2875 美元。
- 按照远期合约以 1.22 的汇率将 1.2875 美元兑换成英镑，得到 $ \frac{1.2875}{1.22} \approx 1.0553 $ 英镑。
- 偿还借款 1.08 英镑。

##### 4. 现金流：
| 时间点    | 操作                                | 美元现金流  | 英镑现金流  |
|-----------|-------------------------------------|-------------|-------------|
| 当期      | 借入 1 英镑                         | $0          | +1 GBP      |
| 当期      | 将 1 英镑兑换为美元                 | +1.25 USD   | -1 GBP      |
| 当期      | 将 1.25 美元存入账户                | $0          | $0          |
| 一年后    | 收到美元存款利息                    | +1.2875 USD | $0          |
| 一年后    | 按远期合约兑换英镑                 | -1.2875 USD | +1.0553 GBP |
| 一年后    | 偿还借款                             | $0          | -1.08 GBP   |
| **净现金流** |                                     | **$0**      | **-0.0247 GBP** |

由于1年后的净现金流是负的，所以这里并不是无风险套利。如果市场存在这样高的远期汇率，我们可以反向操作，即：
- 借入美元，并存入英镑账户，从而无风险套利。

表格如下：

| 时间点    | 操作                                | 美元现金流  | 英镑现金流  |
|-----------|-------------------------------------|-------------|-------------|
| 当期      | 借入 1.25 美元                      | +1.25 USD   | $0          |
| 当期      | 将 1.25 美元兑换为英镑              | -1.25 USD   | +1 GBP      |
| 当期      | 将 1 英镑存入账户                   | $0          | $0          |
| 一年后    | 偿还美元借款                        | -1.2875 USD | $0          |
| 一年后    | 收到英镑存款利息                    | $0          | +1.08 GBP   |
| 一年后    | 按远期合约兑换成美元                | +1.32 USD   | -1.08 GBP   |
| **净现金流** |                                     | **+0.0325 USD** | **$0**      |

这种情况下，净现金流为正，所以投资者可以通过反向套利获得无风险收益。
